### Problem 2

```python
file = open("input.txt", "r") 
lines = file.readlines() 
```

```python
frequency = 0 
```

```python
frequency_vec = []
```

```python
flag = True
while flag:
    for line in lines:
        if frequency in frequency_vec:
            print(frequency)
            flag = False
            break
        frequency_vec.append(frequency) 
        frequency += int(line)
```
