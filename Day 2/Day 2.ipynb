### Problem 1

```python
from collections import Counter

file = open("input.txt", "r")
lines = file.readlines()
```

```python
twos, threes = 0, 0
for line in lines:
    if 2 in Counter(line).values():
        twos += 1
    if 3 in Counter(line).values():
        threes += 1
checksum = twos * threes
print(checksum)
```

### Problem 2

```python
def correct_boxes(box1, box2):
    flag = 0
    for b1, b2 in zip(box1, box2):
        if b1 != b2:
            flag += 1
            char_box1 = b1
        if flag > 1:
            return None
    return print(box1.replace(char_box1, ""))
```

```python
gaussian_sum = len(lines)*(len(lines)+1)/2
print(f"The number of times the function will excute is {gaussian_sum:.0f}.")
```

```python
for i in range(len(lines)):
    for j in range(i+1, len(lines)):
        correct_boxes(lines[i], lines[j])
```
